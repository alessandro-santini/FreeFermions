<center><h1> Sudden Quench </h1></center>

```python
hz = np.ones(64)
hz[32] = -1
H0 = IsingHamiltonian(L=64, Jx=0, Jy=0, hz=hz,pbc=True)
H1 = IsingHamiltonian(L=64, Jx=1, Jy=0, hz=.5,pbc=True)
engine = sudden_quench(H0,H1)
```

```python
t_space = np.linspace(0,20,1001)
n = [np.diagonal(engine.corr.G)]
S =  [engine.corr.compute_entanglement_entropy()]
for t in t_space:
    engine.time_evolve(t)
    n.append(np.diagonal(engine.corr.G))
    S.append(engine.corr.compute_entanglement_entropy())
    
n = np.real_if_close(np.array(n))
S  = np.array(S)
```

```python
plt.imshow(n, aspect='auto', origin='lower',extent=[0,64,0,t_space[-1]])
cb = plt.colorbar()
```

```python
plt.imshow(S, aspect='auto', origin='lower',extent=[1,64,0,t_space[-1]])
cb = plt.colorbar()
```

```python
plt.plot(t_space,S[:-1,32])

plt.xlabel(r"$t$",fontsize=22)
plt.ylabel(r"$S_{L/2}$",fontsize=22)
plt.tick_params(labelsize=18)
```
